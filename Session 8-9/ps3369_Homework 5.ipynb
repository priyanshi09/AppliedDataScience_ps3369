### 1 (10 pts)
Which of the following (select all that apply) is true for the binary classification confusion matrix report:
* a) it reports sizes of each cluster and number of the clusters;
* b) it is a square matrix;
* c) it is symmetrical;
* d) it has dimensions $2\times2$;
* e) it has dimensions $3\times3$;
* f) it reports the quality of classification fit over the test set;
* g) it reports the percentage of correctly classified datapoints over the test set;
* h) it reports the percentage of incorrectly classified datapoints over the test set;

Answer 1)    
a) True:  for the binary classifiication the matrix will be 2*2, which gives number of clusters as 2. For size of each cluster we can calculate it by adding true positives and false positives from confusion matrix.   
b) True: it is a square matrix.    
c) False: It is not necessarily symmetrical, since the transpose of matrix is not always equal to the matrix.    
d) True: For binary classification, the dimension is 2*2 only, including TP, FP, TN and FN as elements.    
e) False: for binary classification the dimension can not be 3*3.     
f) True: it reports the quality  of classification fit over the test set. Since you can calculate accuracy, precision, separability of your model from confusion matrix which gives quality of classification fit over the test set.      
g) True: as it gives you the value of TP, FP, TN, FN which can be used to calcualte the percentage of correctly classified datapoints over the test set.        
h) True: same as g for incorrectly classified datapoints over the test set.   

### 2 (36 pts = 2+4+4+4+7$\times$2+8)
Read the data from https://serv.cusp.nyu.edu/~cq299/ADS2016/Data/HWlogreg2.csv
and then use the first half as the training set, and the second half as the test set.
1. Initiate a logistic regression object for modeling, setting C = 10,000 to disable regularization
2. Run logistic regression on the training set
3. Print out the fitted parameters (including the intercept)
4. Use these parameters to predict on the test set
5. Draw two plots, first one showing your prediction, second one for the actual labels, with decision boundary in both
6. Calculate the confusion matrix to evaluate the model's performance, print out accuracy, recall and precision

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.linear_model import LogisticRegression
%pylab inline
```

```python
data= pd.read_csv("https://serv.cusp.nyu.edu/~cq299/ADS2016/Data/HWlogreg2.csv")
print("the original data length is ", data.shape)
train= data.iloc[:100,:]
test= data.iloc[100:,:]
print("the training set length is ", train.shape)
print("the test set length is ",test.shape)
data.head()
```

```python
fig1 = plt.figure(figsize = (20,6))
plt.suptitle("Quick peek of the data set", fontsize = 18)
ax1, ax2 = fig1.add_subplot(121), fig1.add_subplot(122)
ax1.scatter(train['x1'], train['x2'], lw = 0, s = 80, c = train['y'])
ax2.scatter(test['x1'], test['x2'], lw = 0, s = 80, c = test['y'])
ax1.set(title = 'Traning set')
ax2.set(title = 'Test set')
```

```python
# 1) Initiate a logistic regression object for modeling, setting C = 10,000 to disable regularization
logit = LogisticRegression(C=1e4)

logit.fit(np.c_[train['x1'], train['x2']],train['y'])   # 2) Run logistic regression on the training set

print("The fitted parameters are: ", logit.coef_)  # 3) Print out the fitted parameters (including the intercept)
print("The intercept is ", logit.intercept_)


#apply to the validation sample
labelPred = logit.predict(np.c_[test['x1'], test['x2']])   # 4) Use these parameters to predict on the test set
labelActual = test['y']

# 5) Draw two plots, first one showing your prediction, second one for the actual labels, with decision boundary in both
fig2 = plt.figure(figsize = (20,6))   
plt.suptitle("Logistic Regression's Performance on test set", fontsize = 18)
ax1, ax2 = fig2.add_subplot(121), fig2.add_subplot(122)

#decision boundary
xplot = np.linspace(-5,20,200)
yplot = -(logit.intercept_[0] + logit.coef_[0][0]*xplot)/logit.coef_[0][1]
ax1.plot(xplot,yplot,lw = 2)
ax2.plot(xplot,yplot,lw = 2)

ax1.set(title='Predicted Label')
ax2.set(title='Actual Label')
ax1.scatter(test['x1'], test['x2'], lw = 0, s = 80, c = labelPred)
ax2.scatter(test['x1'], test['x2'], lw = 0, s = 80, c = labelActual)
plt.show()
```

```python
# 6) Calculate the confusion matrix to evaluate the model's performance, print out accuracy, recall and precision

TP = sum((labelPred==1)&(labelActual==1))
FP = sum((labelPred==1)&(labelActual==0))
FN = sum((labelPred==0)&(labelActual==1))
TN = sum((labelPred==0)&(labelActual==0))

#compute precision, accuracy and recall
Acc = 100.0 * (TP+TN)/(TP+TN+FP+FN)
Rec = 100.0 * TP/(TP+FN)
Prec= 100.0 * TP/(TP+FP)
print('Precision = %.2f%%\nAccuracy = %.2f%%\nRecall = %.2f%%'%(Prec,Acc,Rec))
```

```python
print(confusion_matrix(labelPred, labelActual))
print('\nTP = %d, TN = %d, FP = %d, FN = %d'%(TP,TN,FP,FN))
```

All three are quite high, indicating our model's good performance, thanks to the separability of the data.
